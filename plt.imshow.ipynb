```python
import matplotlib.image as img
import matplotlib.pyplot as plt
import numpy as np 
```

```python
array_image = img.imread('Gaurav.jpg')
array_image
```

```python
plt.plot(1,2,2)
plt.title('Original')
plt.imshow(array_image[:])
plt.show()

plt.plot(1,2,1)
plt.title('Mirror Image')
plt.imshow(array_image[:,::-1])
plt.show()

plt.plot(1,2,2)
plt.title('Reflection Image')
plt.imshow(array_image[::-1,::-1])
plt.show()

plt.plot(1,2,2)
plt.title('Pixelated image')
plt.imshow(array_image[::40,::40])
plt.show()

plt.plot(1,2,2)
plt.title('More Pixelated image')
plt.imshow(array_image[::80,::80])
plt.show()
```
